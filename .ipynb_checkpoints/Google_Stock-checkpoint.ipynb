```python
import pandas as pd
import quandl
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = quandl.get('WIKI/GOOGL')

df.head(10)
```

```python
#Creating a dataframe with essential variables

df=df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]


#Defining the relationship between variables

df['HighLow_Percent'] = (df['Adj. High'] - df['Adj. Close']) / df['Adj. Close'] * 100

df['Percent_Change'] = (df['Adj. Close'] - df['Adj. Open']) / df['Adj. Open'] * 100

#Defining the new dataframe

df = df [['Adj. Close', 'HighLow_Percent', 'Percent_Change', 'Adj. Volume']]

df.head(10)
```

```python
#New labels


forecast_col = 'Adj. Close'

#Replacing the NaN values - can be treated as an outlier later 

df.fillna(-99999, inplace=True)

forecast_out = int(math.ceil(0.01*len(df)))
print(forecast_out)


#Using math.ceil to round up values to the nearest whole number 

forecast_out = int(math.ceil(0.01*len(df)))

df['label'] = df[forecast_col].shift(-forecast_out)

df.dropna(inplace=True)

df.head(10)
```

```python
#Defining X and y values


X = np.array(df.drop(['label'], 1))
y = np.array(df['label'])

X = preprocessing.scale(X)

df.dropna(inplace=True)
y = np.array(df['label'])

#print(len(X), len(y))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

df.dropna(inplace=True)

#Linear Regression 
clf = LinearRegression(n_jobs=-1)
clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)

print (accuracy)


```

```python
#SVM - Support vector machine

clf = svm.SVR()

clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)

print (accuracy)
```
